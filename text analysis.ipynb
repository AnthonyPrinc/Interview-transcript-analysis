```python
import numpy as np

# Word counts per turn
word_counts = [21, 55, 49, 34, 32, 36, 38, 45, 35, 53, 34, 22, 37, 25]

# Calculate quartiles and other statistics for turn density
turn_density_stats = {
    'min': np.min(word_counts),
    'first_quartile': np.percentile(word_counts, 25),
    'median': np.median(word_counts),
    'third_quartile': np.percentile(word_counts, 75),
    'max': np.max(word_counts)
}

# Interactivity (turns per minute is constant for all turns)
interactivity = 14 / 5.18  # turns per minute
interactivity_stats = {
    'min': interactivity,
    'first_quartile': interactivity,
    'median': interactivity,
    'third_quartile': interactivity,
    'max': interactivity
}

turn_density_stats, interactivity_stats
```
